```python
ibov_2008 = ibov[ibov.index.year == 2008]
ibov_2008['Close'].plot(figsize=(22,8), label='IBOV')
ibov_2008['Close'].rolling(21).mean().plot(label='MM21')
ibov_2008['Close'].rolling(200).mean().plot(label='MM200')
plt.legend()
```

```python
ibov_periodo_ano = ibov[(ibov.index.year >= 2008) & (ibov.index.year >= 2018)]
ibov_periodo_ano['Close'].plot(figsize=(22,8), label='IBOV')
ibov_periodo_ano['Close'].rolling(21).mean().plot(label='MM21')
ibov_periodo_ano['Close'].rolling(200).mean().plot(label='MM200')
plt.legend()
```

# Fazendo Correlação DOLAR X IBOVESPA

```python
tickers =['^BVSP', 'USDBRL=X']
carteira = web.get_data_yahoo(tickers, start='2007-01-01' )['Close']
carteira.tail()
```

```python
carteira.describe()
```

```python
# Remove os Dados Nulos
carteira = carteira.dropna()
```

```python
# Mudar os nomes das Colunas
carteira.columns = ['Dolar','IBOV']
```

```python
sns.set()
carteira.plot(subplots=True, figsize=(22,8))
```

```python
sns.heatmap(carteira.corr(), annot=True)
```

```python
carteira['Dolar'].rolling(252).corr(carteira['IBOV']).plot(figsize=(22,8))
```

```python
carteira['IBOV_DOLARIZADO'] = (carteira['IBOV'] / carteira['Dolar'])
carteira
```

```python
sns.set()
carteira.plot(subplots=True, figsize=(22,8))
```

```python
dates = pd.date_range('2018', periods=365)
data = pd.DataFrame(carteira, dates, columns=['IBOV_DOLARIZADO', 'IBOV', 'Dolar' ])
data = data.rolling(7)
sns.lineplot(data=data, palette="tab10", linewidth=2.5)
```
